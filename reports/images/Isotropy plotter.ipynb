```python
#  
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def plot_2d_gausian(mean=[0,0],
                    sigma=[1,1],
                    num_samples=2000,
                    figuresize=[10,9],
                    axes=[-3, 3, -3, 3],
                    rot_angle=0,
                    out_dir=False
                   ):
    """
    Plots a 2d gausian distribution centered at mean and sigma for diagonal covrariance matrix
    rot_angle:  Plot (and so covariance matrix) will be rotated by the rot_angle (radians)
    out_path_name: default, False, if str saves image to the given full relative dir name
    """
    
    # Plot scatter 2d normal
    x_in = np.random.normal(mean[0],sigma[0],num_samples)
    y_in = np.random.normal(mean[1],sigma[1],num_samples)
    
    v = np.array([x_in,y_in])

    # build rotation matrix
    cos = np.cos(rot_angle)
    sin = np.sin(rot_angle)
    R = np.array([[cos, -sin],[sin,cos]])
    
    # rotate the vectors
    v_rot = np.dot(R.T,v)
    x = v_rot[0,:]
    y = v_rot[1,:]
    
    plt.figure(figsize=(figuresize[0],figuresize[1]))
    plt.axis(axes)
    plt.scatter(x, y)
    
    if out_dir!=False:
        img_name="%sSigma=%r_Mean=%r_Angle=pi%r_points%r.png" % (out_dir, sigma, mean, rot_angle/np.pi,num_samples)
        plt.savefig(img_name, bbox_inches="tight")
        
    plt.show(block = False)
```

```python
for i in range(1,7):
    factor = 2*i/6
    
    x = plot_2d_gausian(mean=[0,0],
                    sigma=[1,1],
                    num_samples=130,
                    figuresize=[4.,3.6],
                    axes=[-3, 3, -3, 3],
                    rot_angle=np.pi*factor,
                    out_dir="./isotropy_plots/"
                   )
```
